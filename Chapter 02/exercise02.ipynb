Exercise 2.41

```python
from math import atan2

def angle_from_vectors(v1, v2):
    return atan2(v2[1],v2[0]) - atan2(v1[1],v1[0])

print(angle_from_vectors((-1,0), (-3, -1)))
print(angle_from_vectors((-1,-1), (-1, -3)))
print(angle_from_vectors((1,1), (2, 0)))
```

```python
def rotate(angle, vectors):
    polars = [to_polar(v) for v in vectors]
    rotated_polars = [(l, a + angle) for l, a in polars]
    return [to_cartesian(p) for p in rotated_polars]
```
