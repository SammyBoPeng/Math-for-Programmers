```python
import matplotlib.pyplot as plt
```

How to use Matplotlib to draw 3D elements

Coordinates:

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# size
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

plt.show()
```

Points:

```python

fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# size
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

ax.scatter(1, 2, 3, s=80, c='red')
plt.show()
```

Vector:

```python
fig = plt.figure()
ax = fig.add_subplot(projection='3d')

# size
ax.set_xlim(-5, 5)
ax.set_ylim(-5, 5)
ax.set_zlim(-5, 5)

ax.quiver(0, 0, 0, 2, 1, 4)
plt.show()
```
